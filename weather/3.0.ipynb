```python
import pandas as pd
import datetime
from pyecharts import options as opts
from pyecharts.charts import *
from pyecharts.commons.utils import JsCode
```

```python
data = pd.read_csv('广州天气4.0.csv', encoding='utf-8')
data = data.drop(data.index[-1])
data
```

```python
data.info()
```

```python
data[['日期','星期']] = data['日期'].str.split(' ',expand=True,n=1)
```

```python
data
```

```python
data[['最高温度','最低温度']] = data[['最高温度','最低温度']].apply(lambda x: x.str.replace('°',''))
data.head()
```

```python
#计算晴天
data.loc[data['天气'].str.contains('晴'),'晴天吗']='是'
data.fillna('否',inplace=True)
```

```python
#计算雨天
data.loc[data['天气'].str.contains('雨'),'下雨吗']='是'
data.fillna('否',inplace=True)
```

```python
data.sample(10)
```

```python
data['日期'] = pd.to_datetime(data['日期'])
data[['最高温度','最低温度']] = data[['最高温度','最低温度']].astype('int')
```

```python
data['年份'] = data['日期'].dt.year
data['月份'] = data['日期'].dt.month
data['日'] = data['日期'].dt.day
# 预览
data.sample(5)
```

## 时间序列预测

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split
```

```python
data
```

```python
new_data = data.copy()
new_data.set_index('日期', inplace=True)
```

```python
new_data = new_data[['最高温度', '最低温度']]
new_data
```

```python
# 转换成numpy形式
high_train_data = np.asarray(new_data['最高温度'][:'2022-12-31'])
high_test_data = np.asarray(new_data['最高温度']['2022-12-31':])
```

```python
# 转换成numpy形式
low_train_data = np.asarray(new_data['最低温度'][:'2022-12-31'])
low_test_data = np.asarray(new_data['最低温度']['2022-12-31':])
```

```python
# 生成日期范围
dates = pd.date_range(start='2023-06-01', end='2023-09-01', freq='D')

# 创建一个空的 DataFrame，并使用日期范围作为索引
df = pd.DataFrame(index=dates)
df
```

### 最高温度预测

```python
p = 10
d = 0
q = 1

# 训练ARMA模型
model = ARIMA(high_train_data, order=(p, d, q))  # 替换 p, d, q 分别为你选择的阶数
model_fit = model.fit()

# 测试数据
forecast = model_fit.get_forecast(steps=152+93)
predicted_values = forecast.predicted_mean

df['最高温度'] = predicted_values[152:]

# 绘制测试结果
plt.plot(new_data.loc[:'2022-12-31'].index, new_data['最高温度'][:'2022-12-31'], label='Training Data')
plt.plot(new_data.loc['2022-12-31':].index, new_data['最高温度']['2022-12-31':], label='Actual Data')
plt.plot(new_data.loc['2022-12-31':].index, predicted_values[:152], label='Test Data')
plt.legend()
plt.show()
```

### 最低温度预测

```python
p = 50
d = 0
q = 0

# 训练ARMA模型
model = ARIMA(low_train_data, order=(p, d, q))  # 替换 p, d, q 分别为你选择的阶数
model_fit = model.fit()

# 测试数据
forecast = model_fit.get_forecast(steps=152+93)
predicted_values = forecast.predicted_mean

df['最低温度'] = predicted_values[152:]

# 绘制测试结果
plt.plot(new_data.loc[:'2022-12-31'].index, new_data['最低温度'][:'2022-12-31'], label='Training Data')
plt.plot(new_data.loc['2022-12-31':].index, new_data['最低温度']['2022-12-31':], label='Actual Data')
plt.plot(new_data.loc['2022-12-31':].index, predicted_values[:152], label='Test Data')
plt.legend()
plt.show()
```

```python
# 预测值
plt.plot(df.index, df['最高温度'], label='High temperature(forecast)')
plt.plot(df.index, df['最低温度'], label='Low temperature(forecast)')
plt.legend()
plt.tight_layout()
plt.show()
```
