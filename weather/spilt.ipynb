```python
import pandas as pd
```

```python
df = pd.read_csv('gd_weather(true).csv', encoding='utf-8')
df
```

```python
df.info()
```

```python
df['最高温度'] = df['最高温度'].str.replace('°', '').astype(int)
df['最低温度'] = df['最低温度'].str.replace('°', '').astype(int)
df
```

```python
df[['天气', '风向']] = df[['天气', '风向']].astype(str)
df
```

```python
df.info()
```

```python
wea = df['天气'].value_counts().index.tolist()
wea
```

```python
dict_wea = dict(zip(wea, range(len(wea))))
dict_wea
```

```python
df['天气'] = df['天气'].map(dict_wea).astype(str)
df.head()
```

```python
wind = df['风向'].value_counts().index.tolist()
wind
```

```python
dict_wind = dict(zip(wind, range(len(wind))))
dict_wind
```

```python
df['风向'] = df['风向'].map(dict_wind).astype(str)
df.head()
```

```python
df.to_csv('gd_weather.csv', index=False)
```

```python
# 指定保存文件的路径和文件名
file_path1 = 'weather.txt'
file_path2 = 'wind.txt'

# 打开文件并将字典数据写入文件
with open(file_path1, 'w') as file1:
    for key, value in dict_wea.items():
        file1.write(f'{key}: {value}\n')

with open(file_path2, 'w') as file2:
    for key, value in dict_wea.items():
        file2.write(f'{key}: {value}\n')
```
