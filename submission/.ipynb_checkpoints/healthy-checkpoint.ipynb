```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv', encoding='utf-8')
df
```

```python
df.info()
```

```python
# 去掉id这一列
df = df.drop('id', axis=1)
df
```

```python
print(df['Gender'].value_counts())
print(df['Vehicle_Age'].value_counts())
print(df['Vehicle_Damage'].value_counts())
```

```python
# 将Gender的Male设置为0, Female设置为1
df['Gender'] = df['Gender'].map( {'Male': 0, 'Female': 1} ).astype(int)
# 将Vehicle_Age的< 1 Year设置为0, 1-2 Year设置为1, > 2 Years设置为2
df['Vehicle_Age'] = df['Vehicle_Age'].map( {'< 1 Year': 0, '1-2 Year': 1, '> 2 Years': 2} ).astype(int)
# 将Vehicle_Damage的Yes设置为1, NO设置为0
df['Vehicle_Damage'] = df['Vehicle_Damage'].map( {'Yes': 1, 'No': 0} ).astype(int)
df
```
