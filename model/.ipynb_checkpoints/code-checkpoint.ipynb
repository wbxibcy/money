```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
import platform

plt.rcParams['font.family'] = ['SimHei']
plt.rcParams['axes.unicode_minus']=False 
```

```python
data = pd.read_excel("./1(2).xlsx")
data
```

## PCA主成分分析

```python
from sklearn.decomposition import PCA
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
```

```python
# 降维
pca = PCA(n_components='mle')
pca.fit(X)
# 个数
print('保留的主成分个数：' ,pca.n_components_)
# 输出特征值
print(pca.explained_variance_)
# 输出特征向量
print(pca.components_)
# 降维后的数据
X_new = pca.transform(X)
print(f'降维后的数据:\n{X_new}')
```

```python
# 获取主成分得分
scores = pca.transform(X)

plt.bar(range(1, pca.n_components_ + 1), np.abs(scores.mean(axis=0)))
plt.xlabel('Principal Component')
plt.ylabel('Absolute Score')
plt.title('Principal Component Scores (Absolute Values)')
plt.show()
```

### 结论

因为本身特征的`维数`为50, 但是, 降维后的`维数`为47, 降维好像也没什么大用处，直接使用**原始**的特征。

### 归一化

```python
df = data.copy()
new_data = (df - df.min()) / (df.max() - df.min())
new_data
```

```python
X = new_data.iloc[:,1:]
y = new_data.iloc[:,0]
```

```python
print(X)
print(y)
```

### 分割数据集

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
# 随机抽取20%的测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## 网格搜索

### SVR

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
svr = svm.SVR()

param_grid = {"kernel": ['linear', 'rbf', 'poly'],
              "gamma" : [ 0.5,0.4,1,2,3,5], 
              'C' : [950,500,1000,1500]}

svr = GridSearchCV(svr, param_grid=param_grid, scoring='neg_mean_squared_error', cv= 5, n_jobs=-1)
# cv: 交叉验证
# n_jobs: 线程

svr.fit(X_train, y_train) 

print(svr.best_estimator_)
```

```python
svm_reg = svr.best_estimator_
svm_reg.fit(X_train, y_train)
```

```python
# 训练误差
svm_train_mae = mean_absolute_error(svm_reg.predict(X_train), y_train)
svm_train_mse = mean_squared_error(svm_reg.predict(X_train), y_train)
svm_train_r2 = r2_score(svm_reg.predict(X_train), y_train)
print('SVM mean absolute error: ', svm_train_mae)
print('SVM mean_squared_error: ', svm_train_mse)
print('SVM R2: ', svm_train_r2)

# 测试误差
svm_test_mae = mean_absolute_error(svm_reg.predict(X_test), y_test)
svm_test_mse = mean_squared_error(svm_reg.predict(X_test), y_test)
svm_test_r2 = r2_score(svm_reg.predict(X_test), y_test)
print('SVM testing mean absolute error: ', svm_test_mae)
print('SVM testing mean_squared_error: ', svm_test_mse)
print('SVM R2: ', svm_test_r2)
```

```python
svm_y_train_pred = svm_reg.predict(X_train)
plt.figure(figsize=(10,10))
plt.scatter(y_train, svm_y_train_pred)
plt.plot([-0.2, 1.2], [-0.2, 1.2], 'r-')
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
plt.xlabel('y_train')
plt.ylabel('svm_y_train_pred')
plt.savefig("SVR(train).jpg",dpi=300)
plt.show()
```

```python
svm_y_pred = svm_reg.predict(X_test)
plt.figure(figsize=(10,10))
plt.scatter(y_test, svm_y_pred)
plt.plot([-0.2, 1.2], [-0.2, 1.2], 'r-')
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
plt.xlabel('y_test')
plt.ylabel('svm_y_pred')
plt.savefig("SVR(test).jpg",dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(range(len(y_train)),y_train)
plt.scatter(range(len(y_train)),y_train, marker='^')
plt.plot(range(len(y_train)),svm_y_train_pred)
plt.scatter(range(len(y_train)), svm_y_train_pred, marker='o')
plt.xlabel('样本点')
plt.ylabel('数据')
plt.savefig("SVR_train.jpg",dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(range(len(y_test)),y_test)
plt.scatter(range(len(y_test)),y_test, marker='^')
plt.plot(range(len(y_test)),svm_y_pred)
plt.scatter(range(len(y_test)), svm_y_pred, marker='o')
plt.xlabel('样本点')
plt.ylabel('数据')
plt.savefig("SVR_test.jpg",dpi=300)
plt.show()
```

### plsr

```python
from sklearn.cross_decomposition import PLSRegression
```

```python
# 创建PLSR对象，指定要保留的主成分数量（n_components）
# 根据自动降维后的结果，选择47
# 因为PCA分析过了，所以这次不用网格搜索
plsr = PLSRegression(n_components=47)

# 用训练数据拟合PLSR模型
plsr.fit(X_train, y_train)
```

```python
# 训练误差
plsr_train_mae = mean_absolute_error(plsr.predict(X_train), y_train)
plsr_train_mse = mean_squared_error(plsr.predict(X_train), y_train)
plsr_train_r2 = r2_score(plsr.predict(X_train), y_train)
print('PLSR mean absolute error: ', plsr_train_mae)
print('PLSR mean_squared_error: ', plsr_train_mse)
print('PLSR R2: ', plsr_train_r2)

# 测试误差
plsr_test_mae = mean_absolute_error(plsr.predict(X_test), y_test)
plsr_test_mse = mean_squared_error(plsr.predict(X_test), y_test)
plsr_test_r2 = r2_score(plsr.predict(X_test), y_test)
print('PLSR testing mean absolute error: ', plsr_test_mae)
print('PLSR testing mean_squared_error: ', plsr_test_mse)
print('PLSR testing R2: ', plsr_test_r2)
```

```python
plsr_y_train_pred = plsr.predict(X_train)
plt.figure(figsize=(10,10))
plt.scatter(y_train, plsr_y_train_pred)
plt.plot([-0.2, 1.2], [-0.2, 1.2], 'r-')
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
plt.xlabel('y_train')
plt.ylabel('plsr_y_train_pred')
plt.savefig("PLSR(train).jpg",dpi=300)
plt.show()
```
