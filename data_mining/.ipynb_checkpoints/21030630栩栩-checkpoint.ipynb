```python
import numpy as np

# 设置随机种子以保证可复现性
np.random.seed(0)

# 生成50个数据点
num_samples = 50

# 特征数量
num_features = 2

# 生成特征数据
# 特征1 面积/100
# 特征2 距离市中心的公里数(km)
features = np.random.randn(num_samples, num_features)

# 生成随机的目标值（价格）
target = np.random.randn(num_samples)

# 将特征和目标值设置为大于0
features = np.abs(features)
target = np.abs(target)

# 打印生成的数据
for i in range(num_samples):
    print("样本 {}: 特征: {}, 目标值: {}".format(i+1, features[i], target[i]))
```

```python
# 将特征和目标值合并为一个数组
data = np.column_stack((features, target))

# 将数据保存为CSV文件
np.savetxt('data.csv', data, delimiter=',', header='面积/100m^2,距离市中心的公里数,房价', comments='')
```

```python
import pandas as pd

# 从CSV文件读取数据
df = pd.read_csv('data.csv', encoding='gbk')

# 分成x和y
x = pd.DataFrame(df.loc[:, '面积/100m^2':'距离市中心的公里数'])
y = pd.DataFrame(df.loc[:, '房价'])

# 打印数据
print(x)
print(y)
```

```python
y.describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']

# 创建x轴的数据
l = range(len(y))

# 绘制折线图
plt.plot(l, y)
plt.xlabel('样本索引')
plt.ylabel('房价')
plt.title('房价数据折线图')
plt.show()
```

```python
plt.scatter(l, x['面积/100m^2'])
plt.xlabel('样本索引')
plt.ylabel('面积/100m^2')
plt.title('面积/100m^2的数据散点图')
plt.show()
```

```python
plt.scatter(l, x['距离市中心的公里数'])
plt.xlabel('样本索引')
plt.ylabel('距离市中心的公里数')
plt.title('距离市中心的公里数的数据散点图')
plt.show()
```

```python
import warnings
# 忽略所有警告
warnings.filterwarnings("ignore")

from sklearn import svm
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# 划分数据集为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

'''
    核函数选择：SVR模型使用核函数来将数据映射到高维特征空间，从而更好地处理非线性关系。

    正则化参数：SVR模型有一个正则化参数C，它控制了模型对误差的容忍程度。较小的C值会导致更多的误差容忍，可能会产生更简单的模型，而较大的C值会更严格地拟合训练数据。

    epsilon参数：SVR模型还有一个epsilon参数，用于定义在损失函数中对于小于epsilon的误差的惩罚。较大的epsilon值将允许更多的误差，而较小的epsilon值将更严格地控制误差。
'''


# 训练SVR模型
model = svm.SVR(kernel='rbf', C=0.5, epsilon=0.1)
model.fit(X_train, y_train)

# 获取支持向量的特征值
support_vectors = model.support_vectors_

# 获取支持向量的输出值
support_vector_values = np.ravel(model.dual_coef_)

# 绘制支持向量
plt.scatter(support_vectors[:, 0], support_vector_values, c='r', label='支持向量')

'''
    在SVR模型中，支持向量是对模型拟合结果产生影响的重要样本点。通过绘制支持向量的特征值和输出值，可以更直观地观察这些支持向量在模型中的位置。

    输出的图表是一个散点图，横轴表示特征值，纵轴表示输出值。图中用红色的散点表示支持向量。

    支持向量通常位于模型拟合边界附近，它们对于定义决策边界起到重要作用。通过观察图表，可以看到红色散点集中在模型的边界附近，并且在特征值和输出值之间存在一定的关系。这些支持向量表示了模型对于不同特征值对应的输出值的关注点。

'''

plt.xlabel('特征')
plt.ylabel('输出')
plt.title('SVR模型支持向量可视化')
plt.legend()
plt.show()
```

```python
# 在训练集上进行预测
y_train_pred = model.predict(X_train)
print(f"在训练集上的预测值:\n{y_train_pred}")

# 在测试集上进行预测
y_test_pred = model.predict(X_test)
print(f"在测试集上的预测值:\n{y_test_pred}")

# 评估模型性能
print("训练集均方误差（Mean Squared Error）：", mean_squared_error(y_train, y_train_pred))
print("训练集R平方（R2 Score）：", r2_score(y_train, y_train_pred))
print("测试集均方误差（Mean Squared Error）：", mean_squared_error(y_test, y_test_pred))
print("测试集R平方（R2 Score）：", r2_score(y_test, y_test_pred))

# 可视化预测结果
plt.scatter(y_test, y_test_pred)
plt.plot(np.linspace(-0.5, 3, 100), np.linspace(-0.5, 3, 100), color='r', label='y = x')
plt.xlabel('实际值')
plt.ylabel('预测值')
plt.title('SVM回归模型预测结果')
plt.show()
```
